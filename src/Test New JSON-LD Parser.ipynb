```python
def fetch_object_by_predicate_value(nquads, predicate):
    object_values = []
    for _nquad in nquads:
        if _nquad['predicate']['value'] == predicate:
            object_values.append(_nquad['object']['value'])
    return object_values
```

```python
def parse_nquads_by_object(nquads, object_list):
    # create an empty dictionary for storing results
    results = {}
    # loop through each URI in the object_list
    for _object_uri in object_list:
        # use the object URI as the key in the results dict
        results[_object_uri] = []
        # find all target objects in the nquads
        object_ids = fetch_object_by_predicate_value(nquads, _object_uri)
        # loop through each found object
        for _object_id in object_ids:
            object_predicate_dict = fetch_object_predicate_by_subject_value(nquads, _object_id)
            if object_predicate_dict:
                results[_object_uri].append(object_predicate_dict)
        object_predicate_dict2 = fetch_object_predicate_by_subject_value(nquads, _object_uri)
        if object_predicate_dict2:
            results[_object_uri].append(object_predicate_dict2)
    return results
```
